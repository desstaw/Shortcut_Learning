```python
def resize_image(image_path, size=(224, 224)):
    """
    Resize the image to the specified size.

    Args:
    image_path (str): The path to the image to resize.
    size (tuple): The desired size to resize the image to.

    Returns:
    Image: The resized image.
    """
    with Image.open(image_path) as img:
        resized_img = img.resize(size, Image.ANTIALIAS)
        return resized_img

def process_directory(root_dir):
    """
    Traverse all subdirectories in the given directory and resize all images to 224x224.

    Args:
    root_dir (str): The root directory to start processing.
    """
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith(('.jpg', '.jpeg', '.png',)):
                file_path = os.path.join(subdir, file)
                try:
                    resized_img = resize_image(file_path)
                    resized_img.save(file_path)
                    print(f"Resized and saved {file_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

root_dir = "/content/drive/MyDrive/Masterthesis/Datasets/CelebA/raw_224"

# Process the directory
process_directory(root_dir)
```

```python
# Define the paths
source_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/raw_224'
beard_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/data/processed/beard_224'
no_beard_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/data/processed/no_beard_224'

# Load the CSV file into a DataFrame
#df = pd.read_csv(csv_path)

# Ensure the target directories exist
os.makedirs(beard_directory, exist_ok=True)
os.makedirs(no_beard_directory, exist_ok=True)

# Function to copy images based on the filter
def copy_filtered_images(df, source_dir, target_dir, filter_value):
    filtered_df = df[df['No_Beard'] == filter_value]
    for filename in filtered_df['image_id']:
        source_path = os.path.join(source_dir, filename)
        target_path = os.path.join(target_dir, filename)
        if os.path.exists(source_path):  # Check if the source file exists
            shutil.copy2(source_path, target_path)
            print(f"Copied {filename} to {target_dir}")
        else:
            print(f"File {filename} not found in source directory.")

# Copy images with y=0 to beard directory
copy_filtered_images(filtered_df, source_directory, beard_directory, 0)

# Copy images with y=1 to no_beard directory
copy_filtered_images(filtered_df, source_directory, no_beard_directory, 1)
```

```python
beard_patch_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/beard_patch_224'
beard_nopatch_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/beard_nopatch_224'
no_beard_patch_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/no_beard_patch_224'
no_beard_nopatch_directory = '/content/drive/MyDrive/Masterthesis/Datasets/CelebA/no_beard_nopatch_224'


# Ensure the target directory exists
os.makedirs(beard_patch_directory, exist_ok=True)
os.makedirs(beard_nopatch_directory, exist_ok=True)
os.makedirs(no_beard_patch_directory, exist_ok=True)
os.makedirs(no_beard_nopatch_directory, exist_ok=True)



# Function to copy images based on the filter
def copy_filtered_images(df, source_dir, target_dir, y_value, patch_value):
    filtered_df = df[(df['No_Beard'] == y_value) & (df['Black_Hair'] == patch_value)]
    for filename in filtered_df['image_id']:
        source_path = os.path.join(source_dir, filename)
        target_path = os.path.join(target_dir, filename)
        if os.path.exists(source_path):  # Check if the source file exists
            shutil.copy2(source_path, target_path)
            print(f"Copied {filename} to {target_dir}")
        else:
            print(f"File {filename} not found in source directory.")

# Copy images with y=1 and place=0 to beard_patched_224 directory
copy_filtered_images(filtered_df, source_directory, beard_patch_directory, 0, 1)
# Copy images with y=1 and place=1 to beard_nopatch_224 directory
copy_filtered_images(filtered_df, source_directory, beard_nopatch_directory, 0, 0)
# Copy images with y=0 and place=1 to no_beard_patched_224 directory
copy_filtered_images(filtered_df, source_directory, no_beard_patch_directory, 1, 1)
# Copy images with y=0 and place=0 to no_beard_nopatch_224 directory
copy_filtered_images(filtered_df, source_directory, no_beard_nopatch_directory, 1, 0)
```
