# Replace benign and malignant in different paths


```python
# create empty lists to store filenames of benign and malignant images
list_benign_files = []
list_mal_files = []

# iterate over each row in the DataFrame
for index, row in df.iterrows():
    if row['benign_malignant'] == 'benign':
        # append isic_id with ".jpg" to the list of benign files
        list_benign_files.append(row['isic_id'] + ".jpg")
    elif row['benign_malignant'] == 'malignant':
        list_mal_files.append(row['isic_id'] + ".jpg")


print("Benign Files:", list_benign_files)
print("Malignant Files:", list_mal_files)
```

```python
#Restructure the data directory
data_path = '/content/drive/MyDrive/Masterthesis/Datasets/ISIC_224'
img_path = oj('/content/drive/MyDrive/Masterthesis/Datasets/ISIC_224/raw_224')
processed_path = oj('/content/drive/MyDrive/Masterthesis/Datasets/ISIC_224/data/processed')
benign_path = oj('/content/drive/MyDrive/Masterthesis/Datasets/ISIC_224/data/processed/cancer_224')
malignant_path = oj('/content/drive/MyDrive/Masterthesis/Datasets/ISIC_224/data/processed/no_cancer_224')

os.makedirs(processed_path,exist_ok = True)
os.makedirs(benign_path,exist_ok = True)
os.makedirs(malignant_path,exist_ok = True)
```

```python
# Save benign & malignant separately

def save(my_list, my_folder):
    for i,file_name in tqdm(enumerate(my_list)):
        try:
            img = Image.open(oj(img_path, file_name))
            #test = np.asarray(img)
            #img_resized = img.resize((224, 224))
            imwrite(oj(my_folder, file_name), img)
        except:
            print(file_name)

save(list_mal_files, malignant_path)
save(list_benign_files, benign_path)
```
